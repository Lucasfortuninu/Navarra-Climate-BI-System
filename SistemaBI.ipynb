```python
# Funciones para la lectura

def leerId(file):
    idEstacion = 0
    x = file.read(1)
    while(x != ';'):
        idEstacion = idEstacion*10+int(x)
        x = file.read(1)
    return(str(idEstacion))

def leerPueblo(file):
    nombre = ''
    x = file.read(1)
    while(x != ';'):
        nombre = nombre+str(x)
        x = file.read(1)
    return(nombre)

def leerEstacion(file):
    estacion = ''
    x = file.read(1)
    while(x != ';'):
        estacion = estacion+str(x)
        x = file.read(1)
    return(estacion)

def leerComarca(file):
    comarca = ''
    x = file.read(1)
    while(x != ';'):
        comarca = comarca+str(x)
        x = file.read(1)
    return(comarca)
```

```python
# Carga de los CSVs de las estaciones

import requests
import os

# Creo directorio principal si no existe
os.makedirs('datos', exist_ok=True)

def descargar_csv(idEst, pueblo, estacion, anio, ruta):
    # Descargo el archivo CSV y lo guardo en la ruta especificada.
    if anio < 2019 and estacion == 'mapa':
        estacion = 'mapama'
    elif anio >= 2019 and estacion == 'mapama':
        estacion = 'mapa'

    url = (f"http://meteo.navarra.es/_data/datos_estaciones/estacion_{idEst}/datos%20diarios/"
           f"{pueblo.replace(' ', '%20').lower()}%20{estacion}_{anio}.csv")
    fichero = requests.get(url)

    # Creo directorio si no existe
    os.makedirs(ruta, exist_ok=True)

    archivo_csv = os.path.join(ruta, f"{pueblo.replace(' ', '_')}_{anio}.csv")
    with open(archivo_csv, 'wb') as archivo:
        archivo.write(fichero.content)
    
    return archivo_csv

def verificar_csv(archivo_csv):
    # Verifico si el CSV descargado tiene el contenido esperado.
    try:
        with open(archivo_csv, 'rb') as file:
            contenido = file.read()
            if contenido.startswith(b"Fecha-hora;"):
                print(f"Fichero {archivo_csv} creado con éxito")
                return True
            else: # Elimino aquellos csv de pueblos que en ese año no se recogieron datos
                print(f"Fichero {archivo_csv} no válido, eliminando...")
                os.remove(archivo_csv)
                return False
    except FileNotFoundError:
        print(f"Error: El archivo {archivo_csv} no existe")
        return False

# Leo el archivo de pueblos
with open('pueblos_completos.txt', 'r') as pueblos:
    while True:
        idEst = leerId(pueblos)
        pueblo = leerPueblo(pueblos)
        estacion = leerEstacion(pueblos).lower()
        comarca = leerComarca(pueblos)
        ruta = os.path.join('datos', pueblo)

        for anio in range(2006, 2024):
            archivo_csv = descargar_csv(idEst, pueblo, estacion, anio, ruta)
            verificar_csv(archivo_csv)

        # Leo una línea adicional para continuar
        if not pueblos.read(1):  # Salto de línea
            break
```

```python
# Creacion de las tablas de la BBDD

cursor = db.cursor()

# Creo tabla de dimension: LK_comarcas
cursor.execute("""
CREATE TABLE IF NOT EXISTS LK_comarcas (
    ID_comarca INT AUTO_INCREMENT PRIMARY KEY,
    DS_comarca VARCHAR(255) NOT NULL UNIQUE
);
""")



# Creo tabla de dimension: LK_pueblos
cursor.execute("""
CREATE TABLE IF NOT EXISTS LK_pueblos (
    ID_pueblo INT AUTO_INCREMENT PRIMARY KEY,
    DS_pueblo VARCHAR(255) NOT NULL,
    ID_comarca INT NOT NULL,
    CONSTRAINT fk_comarcas_pueblos FOREIGN KEY (ID_comarca) REFERENCES LK_comarcas(ID_comarca) ON DELETE CASCADE ON UPDATE CASCADE
);
""")


# Creo tabla de dimension: LK_tiempo
cursor.execute("""
CREATE TABLE IF NOT EXISTS LK_tiempo (
    ID_tiempo INT AUTO_INCREMENT PRIMARY KEY,
    DS_fecha DATE NOT NULL,
    DS_dia INT,
    DS_mes INT,
    DS_anio INT,
    DS_semana INT,  -- Nueva columna para el número de semana
    CONSTRAINT uq_tiempo_fecha UNIQUE (DS_fecha)
);
""")


# Creo tabla de hechos: DT_temperaturas
cursor.execute("""
CREATE TABLE IF NOT EXISTS DT_temperaturas (
    ID_temperatura INT AUTO_INCREMENT PRIMARY KEY,
    ID_pueblo INT NOT NULL,
    ID_fecha INT NOT NULL,
    IND_tmin FLOAT,
    IND_tmed FLOAT,
    IND_tmax FLOAT,
    IND_precipitacion FLOAT,
    CONSTRAINT fk_temperaturas_pueblos FOREIGN KEY (ID_pueblo) REFERENCES LK_pueblos(ID_pueblo) ON DELETE CASCADE ON UPDATE CASCADE
);
""")

# Confirmo los cambios en la base de datos
db.commit()

cursor.close()
```

```python
# Funciones para poblar la tabla de temperaturas

def saltarFila(csv):
    sep = bytes("\n","utf-8")
    x = csv.read(1)
    while(x != sep):
        x = csv.read(1)
def leerFecha(csv):
    x = csv.read(10)
    x = str(x,"utf-8")
    dia = x[0]+x[1]
    mes = x[3]+x[4]
    anio = x[6]+x[7]+x[8]+x[9]
    fecha = anio + '/' + mes + '/' + dia
    return(fecha, anio, mes, dia)

def saltarHora(csv):
    x = csv.read(10)

def saltarHumedad(csv):
    x = csv.read(5)
    
def leerTemperatura(csv):
    sep = bytes(";","utf-8")
    temp = ""
    x = csv.read(1)
    while(x != sep):
        temp = temp + str(x,"utf-8")
        x = csv.read(1)
    if (temp == ''):
        return None
    return(float(temp.replace(",",".")))
    
def leerTemperatura2(csv):
    sep = bytes(";", "utf-8")
    temp = ""
    x = csv.read(1)
    # Si encuentro `b'\r'`, devuelvo none
    if x == b'\r':
        return None, x
    else:
        # Leo mientras no encuentre `;`
        while x and x != sep and x != b'\r':
            temp += str(x, "utf-8")
            x = csv.read(1)

        # Si no leí nada, el valor es none
        if temp.strip() == '':
            return None, x

    # Convierto el resultado a float reemplazando `,` por `.`
    return (float(temp.replace(",", ".")), x)
 
def saltarDatos(csv):
    salto = bytes("\r","utf-8")
    x = csv.read(1)
    while(x != salto):
        x = csv.read(1)
```

```python
# Evaluación del rendimiento de la BBDD

import mysql.connector
import time
import threading

# Configuración de conexión a la base de datos
db_config = {
    'host': 'eim-srv-mysql.lab.unavarra.es',
    'user': 'aaee_20',
    'password': 'aaee_20',
    'database': 'aaee_20'
}

# Consulta SQL específica
query = """
SELECT 
    DS_fecha AS tiempo,
    SUM(IND_precipitacion) AS precipitacion_anual
FROM 
    DT_temperaturas
INNER JOIN 
    LK_pueblos ON DT_temperaturas.ID_pueblo = LK_pueblos.ID_pueblo
INNER JOIN 
    LK_comarcas ON LK_pueblos.ID_comarca = LK_comarcas.ID_comarca
INNER JOIN 
    LK_tiempo ON DT_temperaturas.ID_fecha = LK_tiempo.ID_tiempo
WHERE 
    LK_comarcas.DS_comarca = 'Pamplona'
GROUP BY 
    DS_anio;
"""

# Función para medir el tiempo de ejecución de una consulta
def measure_query_performance(query):
    try:
        connection = mysql.connector.connect(**db_config)
        cursor = connection.cursor()
        start_time = time.time()
        cursor.execute(query)
        cursor.fetchall()  # Obtén todos los resultados
        end_time = time.time()
        cursor.close()
        connection.close()
        return end_time - start_time  # Duración de la consulta en segundos
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        return None

# Mido el rendimiento
print("Medición de rendimiento de la consulta:")
execution_time = measure_query_performance(query)
if execution_time is not None:
    print(f"Tiempo de ejecución: {execution_time:.4f} segundos\n")

# Estimo las consultas por segundo
if execution_time:
    queries_per_second = 1 / execution_time
    print(f"Consultas por segundo estimadas: {queries_per_second:.2f}")

# Simulo los usuarios concurrentes
def simulate_user_load(query, results):
    execution_time = measure_query_performance(query)
    if execution_time is not None:
        results.append(execution_time)

print("\nSimulación con 100 usuarios concurrentes:")
results = []
threads = []
for i in range(100):  # 100 usuarios concurrentes
    thread = threading.Thread(target=simulate_user_load, args=(query, results))
    threads.append(thread)
    time.sleep(1)  # Simulación de clientes espaciados por 1 segundo
    thread.start()

# Espero a que todos los threads terminen
for thread in threads:
    thread.join()

if results:
    avg_time_concurrent = sum(results) / len(results)
    print(f"Promedio de tiempo por consulta con concurrencia: {avg_time_concurrent:.4f} segundos")
    queries_per_second_concurrent = 1 / avg_time_concurrent
    print(f"Consultas por segundo estimadas con concurrencia: {queries_per_second_concurrent:.2f}")
```
